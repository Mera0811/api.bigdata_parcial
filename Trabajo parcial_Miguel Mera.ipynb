Órdenes de servicios realizadas a través de los Catálogos Electrónicos [Central de Compras Públicas - PERÚ COMPRAS]

El dataset representa las solicitudes de servicios efectuadas mediante el sistema de Catálogos Electrónicos de Acuerdos Marco, gestionado por la entidad oficial de compras públicas, PERÚ COMPRAS. Cada entrada en el dataset corresponde a una solicitud específica de servicio, la cual se ajusta a los criterios de selección y procedimientos operativos establecidos por la Dirección de Acuerdos Marco de PERÚ COMPRAS. Las normativas detalladas en los enlaces proporcionados delinean con precisión dichos criterios y procedimientos, asegurando la transparencia y eficiencia en el proceso de adquisiciones públicas:

```
#Desarrollo del parcial
import requests
from pyspark.sql import SparkSession
from pyspark.sql import Row 
import json
# Conexión al api_url
api_url = 'https://raw.githubusercontent.com/Mera0811/api.bigdata_parcial/main/csvjson.json'
response = requests.get(api_url)
json_data = response.json()
# Sesión Spark
spark = SparkSession.builder.appName("órdenes").getOrCreate()
# Diccionario a lista Row
json_rdd = spark.sparkContext.parallelize(json_data).map(lambda x: Row(**x))
# Dataframe obtenido de la lista Row
dataframe = spark.createDataFrame(json_rdd)
```

```
dataframe.createOrReplaceGlobalTempView("servicios")
```

```
spark.catalog.listTables("global_temp")
```

```
%sql
select * from global_temp.servicios
```
